```python
X = data.drop('charges',axis=1)
y = data['charges']
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler  = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.svm import SVR
model = SVR(kernel='rbf',gamma = 3, C=13500)
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
```

```python
print(y_train_pred)
print(y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
print("mean_absolute_error: ", mean_absolute_error(y_train_pred, y_train))
print("mean_absolute_percentage_error: ", mean_absolute_percentage_error(y_train_pred, y_train))
print("root_mean_squared_error: ", np.sqrt(mean_squared_error(y_train_pred, y_train)))
```

```python
y_test_pred = model.predict(X_test)
```

```python
x = list(range(1, len(y_test)+1))
fig = plt.figure(figsize=(30,30))
ax1 = fig.add_subplot(111)

ax1.scatter(x , y_test , color='red' , label = 'Actual costs')
ax1.scatter(x, y_test_pred , color='green' , label = 'Predicted costs')
```

```python
dif = 10
y_train_acc = []
y_train_pred_acc = []
for index , value in enumerate(y_train):
    y_train_acc.append(1)
    if  y_train_pred[index]<= ((1 + dif/100) * value) and y_train_pred[index] >= ((1 - dif/100) * value): 
        y_train_pred_acc.append(1)
    else:
        y_train_pred_acc.append(0)  
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train_acc,y_train_pred_acc)
```

```python
dif = 10
y_test_acc = []
y_test_pred_acc = []
for index , value in enumerate(y_test):
    y_test_acc.append(1)
    if  y_test_pred[index]<= ((1 + dif/100) * value) and y_test_pred[index] >= ((1 - dif/100) * value):
        y_test_pred_acc.append(1)
    else:
        y_test_pred_acc.append(0)    
```

```python
accuracy_score(y_test_acc,y_test_pred_acc)
```
